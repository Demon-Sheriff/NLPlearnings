```python
#  we can modify spacy to break slangs like gimme into splittings
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme",[
    {ORTH: "gim"},
    {ORTH: "me"}
])

doc = nlp("gimme double cheese burger with extra hot sauce")
tokens = [token for token in doc]
print(tokens)
```

```python
doc = nlp("Monkey D. Luffy ate 100$ of pizza. Then he said I can do this all day.")
```

```python
dir(doc[0])
for token in doc:
    # pos is part of speech
    print(token," | ",token.pos_, " | ", token.lemma_)
```

```python
doc = nlp("I want to make a contestBot which will give you reminders and contest details from LeetCode, Codeforces, Codechef etc.")

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ", spacy.explain(ent.label_))
```
