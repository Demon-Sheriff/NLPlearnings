```python
# Classify the ham ones as 1 and spam ones as 0 for a new column Spam
data_spam['Spam'] = [1 if data_spam.iloc[i]['Category'] == 'spam' else 0 for i in range(len(data_spam))]
data_spam['NotSpam'] = [1 if data_spam.iloc[i]['Category'] == 'ham' else 0 for i in range(len(data_spam))]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_spam.Message,data_spam.Spam,test_size=0.25)
```

```python
# We are going to use multinomial naive bayes
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:5]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,Y_train)
```

```python
# lets take some sample emails
emails = [
          'Hey mohan, can we get together to watch football game tomorrow?',
          'Upto 20% discount on parking, exclusive offer just for you. Dont miss thi reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, Y_test)
```

```python
# Lets do the same thing that we did above using the sklearn pipeline

from sklearn.pipeline import Pipeline
clf = Pipeline([('vectorizer',CountVectorizer()),
                ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train, Y_train)
```

```python
clf.score(X_test, Y_test)
```
