# Stemming using NLTK

```python
from nltk.stem import PorterStemmer, SnowballStemmer
stemmer = PorterStemmer()
```

```python
words = ["eating","eats","eat","ate","adjustable","rafting","ability","meeting"]
```

```python
for word in words:
    print(word, " | ", stemmer.stem(word))
    
for word in words:
    print(word,end=" ")
```

# Lemmatization using spacy

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("eating eats eat ate adjustable rafting ability meeting better")

for token in doc:
    print(token," | ", token.lemma_, " | ", token.lemma)
# note that lemma_ is the actual lemma as a string of the token 
# whereas the lemma is the hashed unique integer stored in spacy for each lemma string
```

```python
doc = nlp("Mando talked for 3 hours although talking isn't his thing, he thinks he became talkative")

for token in doc:
    print(token," | ", token.lemma_, " | ", token.lemma)
```

```python
nlp.pipe_names
```

```python
# let's check out some slangs
doc = nlp("Bro, do you wanna go ? Brah, don't say no! I am exhausted")

for token in doc:
    print(token," | ", token.lemma_, " | ", token.lemma)
# the slangs dont get converted into the correct base words / lemmas
```

```python
# so let's customize our pipeline to include the above slangs
ar = nlp.get_pipe('attribute_ruler')

ar.add([[{"TEXT":"Bro"}],[{"TEXT":"Brah"}]], {"LEMMA": "Brother"})
doc = nlp("Bro, do you wanna go ? Brah, don't say no! I am exhausted")
```

```python
# let's check it now
for token in doc:
    print(token," | ", token.lemma_)
```
